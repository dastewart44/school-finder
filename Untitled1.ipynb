```python
fake_data = 'https://storage.googleapis.com/school_finder_bucket/fake_data_avgs.csv'
df = pd.read_csv(fake_data)
df
```

```python
import plotly.express as px

fig = px.bar(df, x='school_id', y='student_count', title='Student Counts by School ID')
fig.show()
```

```python
fig = px.scatter(df, x='start_gpa_avg', y='end_gpa_avg', title='Starting GPA vs. Ending GPA',
                 hover_data=['school_id', 'student_count'])
fig.show()
```

```python
fig = px.pie(df, names=['pct_asian', 'pct_black', 'pct_hispanic', 'pct_white', 'pct_other'],
             title='Ethnicity Percentages', hole=0.4)
fig.show()
```
